```python
# CÃ¢u 7
# CÃ´ng thá»©c tÃ­nh: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
# df[tenconmoi] = df[]
```

```python
for i in ["1", "2", "6"]:
    print("TBM" + i)
```

```python
for i in ["1", "2", "6"]:
    df["TBM" + i] = (df["T" +i]*2 + df["L" +i] + df["H" +i] + df["S" +i] + df["V" +i]*2 + df["X" +i] + df["D" +i] + df["N" +i]) / 10
```

```python
df.rename(columns={"TBM6":"TBM3"}, inplace= True)
```

```python
df.head()
```

```python
# CÃ¢u 8: Táº¡o cÃ¡c biáº¿n xáº¿p loáº¡i XL1, XL2 vÃ  XL3 dá»±a trÃªn TBM1, TBM2 vÃ  TBM3
def xep_loai(tbm):
    if tbm < 5.0:
        return "Y"
    elif tbm < 6.5:
        return "TB"
    elif tbm < 8.0:
        return "K"
    elif tbm < 9.0:
        return "G"
    else:
        return "XS"
```

```python
df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)
```

```python
df.head()
```

```python
# CÃ¢u 9: Thá»±c hiá»‡n viá»‡c chuyá»ƒn Ä‘á»•i tá»« thang Ä‘iá»ƒm 10 sang thang Ä‘iá»ƒm 4 báº±ng phÆ°Æ¡ng phÃ¡p Min-Max Normalization.
def min_max_normalization(x, min_x=0, max_x=10, min_y=0, max_y=4):
    return (x - min_x) / (max_x - min_x) * (max_y - min_y) + min_y
```

```python
# Táº¡o cÃ¡c biáº¿n US_TBM1, US_TBM2 vÃ  US_TBM3
df["US_TBM1"] = df["TBM1"].apply(min_max_normalization)
df["US_TBM2"] = df["TBM2"].apply(min_max_normalization)
df["US_TBM3"] = df["TBM3"].apply(min_max_normalization)
```

```python
df.head()
```

```python
# CÃ¢u 10:
# TÃ­nh Ä‘iá»ƒm xÃ©t tuyá»ƒn cho cÃ¡c khá»‘i A, A1, B vÃ  cÃ¡c khá»‘i khÃ¡c
def tinh_diem(row):
    if row['KT'] in ['A', 'A1']:
        diem_xet_tuyen = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif row['KT'] == 'B':
        diem_xet_tuyen = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        diem_xet_tuyen = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return diem_xet_tuyen
```

```python
# Táº¡o biáº¿n Ä‘iá»ƒm xÃ©t tuyá»ƒn
df['diemxettuyen'] = df.apply(tinh_diem, axis=1)
```

```python
# Táº¡o biáº¿n káº¿t quáº£ xÃ©t tuyá»ƒn (KQXT)
df['KQXT'] = df['diemxettuyen'].apply(lambda x: 1 if x >= 5.0 else 0)
```
