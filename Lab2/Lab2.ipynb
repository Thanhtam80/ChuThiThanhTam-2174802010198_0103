## Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối 
1. Hãy mô tả và khảo sát phân phối cho biến T1

```python
t1_desc = df['T1'].describe() 
# Tính toán mô tả thống kê của cột 'T1'
```

```python
print("Mô tả độ tập trung và phân tán của dữ liệu T1:")
t1_desc # xuất thống kê
```

```python
# Vẽ biểu đồ Box-Plot cho cột 'T1'
plt.figure(figsize=(10, 6)) # Tạo một khung hình với kích thước 10x6 inch
sns.boxplot(x=df['T1']) # Vẽ biểu đồ Box-Plot cho cột 'T1'
plt.title('Box-Plot của T1') # đặt tiêu đề
plt.xlabel('Điểm T1') # đặt tên trục x
plt.show() # Hiển thị biểu đồ
```

```python
skewness = df['T1'].skew() # Tính toán độ lệch của 'T1'
kurtosis = df['T1'].kurtosis() # Tính toán độ nhọn của 'T1'
```

```python
print(f"Độ lệch (Skewness): {skewness}") # xuất độ lệch
print(f"Độ nhọn (Kurtosis): {kurtosis}") # xuất độ nhọn
```

```python
plt.figure(figsize=(10, 6)) # Tạo một khung hình với kích thước 10x6 inch
sns.histplot(df['T1'], bins=10, kde=True)
# Vẽ biểu đồ Histogram với 10 bins và biểu đồ KDE
plt.title('Histogram của T1') # Đặt tiêu đề cho biểu đồ
plt.xlabel('T1') # Đặt tên trục x
plt.ylabel('Tần suất') # Đặt tên trục y
plt.show() # Hiển thị biểu đồ
```

```python
plt.figure(figsize=(10, 6)) # Tạo một khung hình với kích thước 10x6 inch
stats.probplot(df['T1'], dist="norm", plot=plt) # Vẽ biểu đồ QQ-Plot để kiểm tra độ chuẩn của phân phối 'T1'
plt.title('QQ-Plot của T1') # Đặt tiêu đề cho biểu đồ
plt.show() # Hiển thị biểu đồ
```

```python
print(f"Trung bình của T1 là: {t1_desc['mean']}")
print(f"Trung vị của T1 là: {t1_desc['50%']}")
print(f"Độ lệch (skewness) của T1 là: {skewness} cho thấy phân phối lệch trái khi giá trị âm")
print(f"Độ nhọn (kurtosis) của T1 là: {kurtosis} cho thấy phân phối có mức độ nhọn thấp hơn phân phối chuẩn khi giá trị < 0")
```
