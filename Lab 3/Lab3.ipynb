Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn "kg"

```python
weight= df['Weight'] # Gán cột Weight vào biến weight

for i in range(0, len(weight)): # Gọi vòng lặp for lặp qua từng phần tử trong cột weight
    x = str(weight[i]) # Chuyển giá trị của phần tử weight[i] thành chuỗi
    if "lbs" in x[-3:]: # Nếu chuỗi x có chứa lbs ở ba ký tự cuối
        x = x[:-3:]# Loại bỏ lbs khỏi chuỗi x
        float_x = float(x) # Chuyển đổi x thành số thực
        y = int(float_x/2.2) # Chuyển đổi đơn vị từ pound sang kilogram bằng cách chia cho 2.2
        y = str(y) + "kgs" # Chuyển đổi kết quả sang chuỗi, thêm "kgs"
        weight[i] = y # Gán giá trị đã chuyển đổi lại cho phần tử tương ứng
```

11. Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp.

```python
df['PulseRate']= df['PulseRate'].replace({'-' : '0'}, regex=True) 
# Thay thế ký tự '-' bằng 0 trong cột PulseRate
df["PulseRate"].fillna('0', inplace =True) # Điền giá trị 0 cho giá trị bị thiếu trong cột PulseRate
```

```python
# Tính toán trung bình của các giá trị huyết áp của nhóm giới tính
def avgPulseRateBySex(df, sex): # Hàm để tính trung bình PulseRate theo sex
    sumPulseRate = 0 # Khởi tạo tổng giá trị PulseRate bằng 0
    nsumPulseRate = 0 # Khởi tạo số lượng giá trị PulseRate bằng 0
    for i, value in enumerate(df['Sex']): # Lặp qua từng giá trị trong cột Sex
        if value == sex: # Nếu giá trị giới tính của hàng hiện tại bằng sex
            try:
                if df['PulseRate'][i] != '0': # Nếu giá trị PulseRate khác 0
                    sumPulseRate += int(df['PulseRate'].values[i]) # Cộng giá trị PulseRate vào sumPulseRate
                    nsumPulseRate += 1  # Tăng số lượng các giá trị thêm 1
            except:
                pass 
    if nsumPulseRate > 0: # Nếu có ít nhất một giá trị PulseRate
        return str(sumPulseRate / nsumPulseRate) # Trả về giá trị trung bình của PulseRate dạng chuỗi
    else: # nếu không có
        return 0 # Trả về 0
```

```python
# Tính toán trung bình huyết áp của các giá trị dữ liệu
def avgPulseRateOverall(df): # Hàm để tính trung bình PulseRate cho dữ liệu df
    sumPulseRate = 0 # Khởi tạo tổng giá trị PulseRate bằng 0
    nsumPulseRate = 0 # Khởi tạo số lượng giá trị PulseRate bằng 0
    for i, value in enumerate(df['PulseRate']): # Lặp qua từng giá trị trong cột PulseRate
        try:
            if value != '0': # Nếu giá trị PulseRate khác 0
                sumPulseRate += int(value) # Cộng giá trị PulseRate vào sumPulseRate
                nsumPulseRate += 1  # Tăng số lượng các giá trị thêm 1
        except:
            pass
    if nsumPulseRate > 0: # Nếu có ít nhất một giá trị PulseRate
        return str(sumPulseRate / nsumPulseRate)  # Trả về giá trị trung bình của PulseRate dạng chuỗi
    else: # nếu không có
        return 0 # Trả về 0
```

```python
def avgPulserate(df, firstname): # Hàm để tính trung bình PulseRate theo firstname
    sumPulserate = 0 # Khởi tạo tổng giá trị PulseRate bằng 0
    nsumPulserate = 0 # Khởi tạo số lượng giá trị PulseRate bằng 0
    for i, value in enumerate(df['Firstname']): # Lặp qua từng giá trị trong cột Firstname
        if value == firstname: # Nếu giá trị Firstname của hàng hiện tại bằng với firstname 
            try:
                if df['PulseRate'][i] !='0': # Nếu giá trị PulseRate khác 0
                    sumPulserate += int(df['PulseRate'].values[i]) # Cộng giá trị PulseRate vào sumPulseRate
                    nsumPulserate += 1 # Tăng số lượng các giá trị thêm 1
            except:
                pass
    if nsumPulserate > 0: # Nếu có ít nhất một giá trị PulseRate
        return str(sumPulserate/nsumPulserate)  # Trả về giá trị trung bình của PulseRate dạng chuỗi
    else: # nếu không có
        return 0 # Trả về 0
```

```python
def isValidIndex1(index, nrows): # Hàm kiểm tra chỉ số hợp lệ trong phạm vi df
    return index - 1 > 0 & index + 1 < nrows # Trả về True nếu chỉ số nằm trong phạm vi hợp lệ, ngược lại trả về False
```

```python
def isValidIndex2(index, nrows): # Hàm kiểm tra chỉ số hợp lệ trong phạm vi df
  return index + 1 < nrows & index + 2 < nrows # Trả về True nếu chỉ số nằm trong phạm vi hợp lệ, ngược lại trả về False
```

```python
def isPulseRateNotNull1(index, df): # Hàm kiểm tra giá trị PulseRate không phải 0
  return str(df['PulseRate'].values[index-1])!= '0' and str(df['PulseRate'].values[index+1])!= '0' # Trả về True nếu giá trị PulseRate của hai hàng xung quanh không phải 0, ngược lại trả về False
```

```python
def isPulseRateNotNull2(index, df): # Hàm kiểm tra giá trị PulseRate không phải 0
  return str(df['PulseRate'].values[index+1])!= '0' and str(df['PulseRate'].values[index +2])!= '0' # Trả về True nếu giá trị PulseRate của hai hàng kế tiếp không phải 0, ngược lại trả về False
```

```python
def isInGroup1(df, index, firstname): #  Hàm kiểm tra giá trị Firstname của hai hàng xung quanh có giống nhau không
  return df['Firstname'].values[index-1]== firstname and df['Firstname'].values[index +1]== firstname # ^rả về True nếu giá trị Firstname của hai hàng xung quanh giống nhau, ngược lại trả về False
```

```python
def isInGroup2(df, index, firstname): # Hàm kiểm tra giá trị Firstname của hai hàng kế tiếp có giống nhau không
  return df['Firstname'].values[index+1]== firstname and df['Firstname'].values[index +2]== firstname # Trả về True nếu giá trị Firstname của hai hàng kế tiếp giống nhau, ngược lại trả về False
```

```python
def updatePulseRatena(index, df): # Hàm cập nhật giá trị PulseRate
    nrows = len(df['PulseRate']) - 2 # Lấy số lượng hàng trong df trừ 2
    newvalue = '' # Khởi tạo giá trị mới rỗng
    firstname = df['Firstname'].values[index] # Lấy giá trị Firstname của hàng hiện tại
    sex = df['Sex'].values[index] # Lấy giá trị Sex của hàng hiện tại
    try:
        if isValidIndex1(index, nrows) and isPulseRateNotNull1(index, df) and isInGroup1(df, index, firstname): # Nếu chỉ số hợp lệ và giá trị PulseRate của hai hàng xung quanh không phải 0 và giá trị Firstname của hai hàng xung quanh giống nhau
            try:
                newvalue = (int(df['PulseRate'].values[index - 1]) + int(df['PulseRate'].values[index + 1])) / 2 # Trung bình giá trị PulseRate của 2 hàng xung quanh
            except:
                newvalue = '0' # Gán giá trị newvalue là 0
        elif isValidIndex2(index, nrows) and isPulseRateNotNull2(index, df) and isInGroup2(df, index, firstname): # Nếu chỉ số hợp lệ và giá trị PulseRate của hai hàng kế tiếp không phải 0 và giá trị Firstname của hai hàng kế tiếp giống nhau
            try:
                newvalue = (int(df['PulseRate'].values[index + 1]) + int(df['PulseRate'].values[index + 2])) / 2 # Trung bình giá trị PulseRate của 2 hàng kế tiếp
            except:
                newvalue = '0' # Gán giá trị newvalue là 0
        else:
            newvalue = avgPulserate(df, firstname) # Lấy trung bình giá trị PulseRate của người đó
            if newvalue == '0': # Nếu giá trị trung bình Pulserate của người đó là 0
                newvalue = avgPulseRateBySex(df, sex) # Lấy trung bình giá trị PulseRate theo sex
                if newvalue == '0': # Nếu giá trị trung bình sex của người đó là 0
                    newvalue = avgPulseRateOverall(df) # Lấy trung bình giá trị PulseRate
    except:
        newvalue = avgPulserate(df, firstname) # Lấy trung bình giá trị PulseRate của người đó
        if newvalue == '0': # Nếu giá trị trung bình Pulserate của người đó là 0
            newvalue = avgPulseRateBySex(df, sex) # Lấy trung bình giá trị PulseRate theo sex
            if newvalue == '0': # Nếu giá trị trung bình sex của người đó là 0
                newvalue = avgPulseRateOverall(df) # Lấy trung bình giá trị PulseRate
    return newvalue # Trả về giá trị newvalue
```

```python
pulserateupdate = [] # Khởi tạo danh sách rỗng để lưu các giá trị PulseRate được cập nhật
for index, value in enumerate(df["PulseRate"]): # Lặp qua từng giá trị và chỉ số trong cột PulseRate
  if value =='0' or value == '': # Nếu giá trị PulseRate = 0 hoặc bị thiếu
    newvalue = updatePulseRatena(index,df) # Cập nhật giá trị PulseRate
    pulserateupdate.append(newvalue) # Thêm giá trị PulseRate được cập nhật vào danh sách pulserateupdate
  else:
    pulserateupdate.append(value) # Thêm giá trị PulseRate hiện tại vào danh sách pulserateupdate
```

```python
df["PulseRateUpdate"] = pulserateupdate # Thêm cột PulseRateUpdate vào dataframe
df # Hiển thị dataframe
```
