Câu 12: Thay thế giá trị cột Age lớn hơn 30 bằng giá trị trung bình cột đó

```python
TB_Age = df['Age'].mean() # lấy trung bình cột Age
```

```python
df.loc[(df['Age'] > 30), 'Age_2']= TB_Age # Thay thế giá trị cột Age lớn hơn 30 bằng giá trị trung bình cột đó
```

```python
df # hiển thị dataframe
```

Câu 13: Chuẩn hóa (normalize) cột Age về khoảng giá trị từ 0 đến 1

```python
df['Chuanhoa_Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min()) 
df # hiển thị dataframe
```

Câu 14: Tạo cột mới Age_group phân loại tuối thành Young, Middle-aged, Old dựa trên giá trị cột Age

```python
def Age_group(age):
    if age < 30:
        return 'Young'
    elif age > 60:
        return 'Old'
    else:
        return 'Middle-aged'
```

```python
df['Age_group'] = df['Age'].apply(Age_group)
df # hiển thị dataframe
```

Câu 15: Tính toán tỷ lệ phần trăm thay đổi (percentage change) cột Salary

```python
df['Percentage_change'] = df['Salary'].pct_change() * 100
df # hiển thị dataframe
```
